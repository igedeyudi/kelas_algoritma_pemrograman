<h2 style="color: #142d4c; background: #9fd3c7; padding: 10px; border-radius:10px"> Tipe Data </h2>

Tipe data paling dasar yang digunakan untuk merepresentasikan jenis dari sebuah nilai dalam bahasa pemrograman dikenal sebagai tipe data fundamental atau primitif (primitive data type). Tipe data seperti itu mengandung nilai data yang murni dan sederhana. 

<h3 style="color: #142d4c; background: #ececec; padding: 10px; border-radius:10px"> Tipe Data Primitif </h3>

Bahasa pemrograman Python memiliki empat tipe data primitif atau fundamental, yaitu integer, float, boolean, dan string.
1. **int**      : Bilangan bulat dari -∞ hingga+∞ adalah int. contohnya adalah 100, 200, -20, 291, -182
2. **float**    : Berbeda dengan int, float atau sering disebut bilangan riil adalah bilangan yang memiliki pecahan. contohnya adalah 100.21, 22.00, -1.20, 4.291, -1.182.
3. **boolean**  : Boolean adalah bilangan yang hanya memiliki dua kemungkinan nilai yakni True atau False. 
3. **str**      : String digunakan untuk merepresentasikan nilai yang berbentuk teks. Untuk mendefinisikan nilai teks digunakan tanda single atau double quotation seperti dicontohkan sebelumnya.

<h3 style="color: #142d4c; background: #ececec; padding: 10px; border-radius:10px"> <i>Casting</i> Tipe Data </h3>

Kita dapat memaksakan untuk membuat sebuah nilai. Hal ini dapat dilakukan dengan casting. 
- int() - membuat bilangan integer dari literal integer, literal float (dengan menghapus semua desimal), atau string literal (dengan syarat string mewakili bilangan bulat)
- float() - membuat angka float dari literal integer, literal float, atau string literal (menyediakan string yang mewakili float atau integer)
- str() - membuat string dari berbagai tipe data, termasuk string, literal integer, dan literal float


```python
int("5") # cast str menjadi int
str(203) # cast int menjadi str
float(203) # cast int menjadi float
```

<h2 style="color: #142d4c; background: #9fd3c7; padding: 10px; border-radius:10px"> Variabel </h2>

Variable adalah objek yang digunakan untuk menampung data atau dengan kata lain mereferensikan ke sebuah data atau nilai tertentu. Seperti gambar di bawah ini Variable x adalah sebuah objek yang direferensikan ke sebuah nilai 7 yang bertipe data int. Namun sesuai namanya nilai yang direpresentasikan sebuah variable dapat berubah.



<h3 style="color: #142d4c; background: #ececec; padding: 10px; border-radius:10px"> Mendefinisikan Variabel </h3>

Untuk mendefinisikan variable di python, caranya yakni buat sebuah identifier diikuti dengan assignment operator (=) dan diikuti dengan nilainya. Kita tidak perlu mendefinisikan tipe data Ketika membuat sebuah variable karena Python akan secara otomatis mengenali berdasarkan nilai yang direferensikan. Bahkan, tipe data sebuah variable dapat berubah jika kita merubah isinya kemudian.

```python
# Mendefinisikan variable
harga = 900
qty = 20

# Cara lain mendefinisikan variable
harga, qty = 900, 20

# Tipe data variable dapat berubah ketika kita mengganti nilainya
harga = "mahal"
```

<h3 style="color: #142d4c; background: #ececec; padding: 10px; border-radius:10px"> Mengakses Variabel </h3>

Setelah didefinisikan, sebuah variable dapat dipanggil atau digunakan sewaktu-waktu jika diperlukan

```python
# Mendefinisikan variable
harga = 900
qty = 20

# Mengakses variable
total = harga * qty
```

<h3 style="color: #142d4c; background: #ececec; padding: 10px; border-radius:10px"> Menghapus Variabel </h3>

Untuk menghpus variable gunakan keyword del diikuti dengan nama variable dari memori seperti contoh berikut:

```python
# menghapus variable
del harga
del qty
```
